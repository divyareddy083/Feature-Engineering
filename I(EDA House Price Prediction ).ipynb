#### Replacing missing values

```python
#Categorical features with nan values
features_with_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>0 and dataset[feature].dtypes=='O']
for feature in features_with_nan:
    print('{} as {}% of missing values'.format(feature,np.round(dataset[feature].isnull().mean(),3)))
```

```python
#Replacing the missing values in categorical features with a new label
def replace_cat_feature(dataset,features_with_nan):
    data=dataset.copy()
    data[features_with_nan]=data[features_with_nan].fillna('Missing')
    return data

dataset=replace_cat_feature(dataset,features_with_nan)

dataset[features_with_nan].isnull().sum()
```

```python
dataset.head()
```

```python
#Numerical features with nan values
numerical_features_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>0 and dataset[feature].dtypes!='O']
for feature in numerical_features_nan:
    print('{} as {}% of missing values in it'.format(feature,np.round(dataset[feature].isnull().mean(),3)))
```

```python
## Replacing the Missing Values in numerical features
for feature in numerical_features_nan:
    ## Let's replace by using median since there are outliers
    median_value=dataset[feature].median()
    ## Lets create another variable to capture nan values 
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
    
dataset[numerical_features_nan].isnull().sum()
```

```python
dataset.tail()
```

#### Temporal Variables (Date Time Variables)

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature]=dataset['YrSold']-dataset[feature]
```

```python
dataset[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'YrSold']].head()
```

##### Since the numerical variables are skewed we will perform log normal distribution

```python
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']
for feature in num_features:
    dataset[feature]=np.log(dataset[feature])
```

```python
dataset.head()
```

#### Handling Rare Categorical Feature

We will remove categorical variables that are present less than 1% of the observations

```python
categorical_features
```

```python
for feature in categorical_features:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(20)
```

```python
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

#### Apply Feature Selection

```python
# first, I specify the Lasso Regression model, and I
# select a suitable alpha (equivalent of penalty).
# The bigger the alpha the less features that will be selected.

# Then I use the selectFromModel object from sklearn, which
# will select the features which coefficients are non-zero

# remember to set the seed, the random state in the below function
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0)) 
feature_sel_model.fit(X_train, y_train)
```

```python
feature_sel_model.get_support()
```

```python
# let's print the number of total and selected features

# this is how we can make a list of the selected features
selected_feat = X_train.columns[(feature_sel_model.get_support())]

# let's print some stats
print('total features: {}'.format((X_train.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrink to zero: {}'.format(np.sum(feature_sel_model.estimator_.coef_== 0))) #this result should be 80 
```
